```python
# Install transformers and torch if not already installed
#!pip install transformers 
#!pip install torch

import pandas as pd
import torch
from transformers import BertTokenizer, BertModel
```

```python
# Load your dataset
# Assuming you have a CSV file with a column named "Details"
data = pd.read_csv("Details.csv")

# Load the BERT model and tokenizer
model_name = "bert-base-uncased"  # You can choose other models like "distilbert-base-uncased" for faster results
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# Function to get BERT embeddings for each text entry
def get_bert_embedding(text):
    # Tokenize the input text and convert to tensor
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    with torch.no_grad():  # Disables gradient calculation
        outputs = model(**inputs)
    # Get the embeddings for the [CLS] token (first token), which represents the whole sequence
    embeddings = outputs.last_hidden_state[:, 0, :].squeeze()
    return embeddings.numpy()

# Apply the embedding function to each row in the "Details" column
data['BERT_Embeddings'] = data['Details'].apply(get_bert_embedding)

# Convert embeddings to separate columns
# Create a new DataFrame by expanding the BERT embedding column
embedding_df = pd.DataFrame(data['BERT_Embeddings'].tolist(), index=data.index)

# Rename columns to something meaningful, like "Feature_1", "Feature_2", ...
embedding_df.columns = [f"Feature_{i+1}" for i in range(embedding_df.shape[1])]

# Concatenate the original data with the new embedding features
result = pd.concat([data.drop(columns=['BERT_Embeddings']), embedding_df], axis=1)

# Save the result to a new CSV file
result.to_csv("dataset_with_bert_features.csv", index=False)

# Display the resulting DataFrame
print(result.head())
```
