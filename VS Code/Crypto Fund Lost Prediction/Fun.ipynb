```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = ['2000', '2004', '2008', '2012', '2016', '2020']
democratic_counts = [150, 250, 350, 400, 300, 500]
republican_counts = [200, 600, 1100, 1300, 1500, 1800]

# Bar positions
y_positions = np.arange(len(years))

# Plot
plt.figure(figsize=(8, 6))
plt.barh(y_positions, democratic_counts, color='skyblue', label='Democratic')
plt.barh(y_positions, republican_counts, left=democratic_counts, color='lightcoral', label='Republican')

# Labels and Title
plt.ylabel('Year')
plt.xlabel('Counties')
plt.title('Americans Are More Reliant Than Ever on Government Aid', fontsize=20, wrap=True)
plt.yticks(y_positions, years)
plt.xticks(np.arange(0, 2250, step=250))
plt.legend()

# Notes and Sources
plt.text(-600, -1.5, 'Note: Significant assistance equates to 25% or more of a county’s total personal income. *2020 county total was bolstered by temporary Covid-19-related spending.\nSources: Economic Innovation Group; U.S. Election Atlas; Bureau of Economic Analysis', 
         fontsize=8, ha='left', va='top', wrap=True)

# Display the plot
plt.show()
```
