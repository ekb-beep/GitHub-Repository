```python

import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

# Using scatter_matrix to create the matrix of scatter plots
columns_list = ['Top10perc', 'Apps', 'Enroll']
scatter_matrix(college[columns_list], alpha=0.7, diagonal='kde')

# Show the plot
plt.show()
```

```python
# Creating a side-by-side boxplot
college.boxplot(by='Private', column='Outstate', grid=False)

# Adding titles and labels
plt.title('Side-by-Side Boxplot')
plt.suptitle('')  # Suppress the default title
plt.xlabel('Private')
plt.ylabel('Outstate')

# Show the plot
plt.show()
```

```python
college['Elite'] = pd.cut(college['Top10perc'],
[0,50,100],
labels=['No', 'Yes'])
```

```python
college['Elite'].value_counts
college['Elite'].describe()
```

```python
# Creating a side-by-side boxplot
college.boxplot(by='Elite', column='Outstate', grid=False, figsize=(7, 5))

# Adding titles and labels
plt.title('Side-by-Side Boxplot')
plt.suptitle('')  # Suppress the default title
plt.xlabel('Elite')
plt.ylabel('Outstate')

# Show the plot
plt.show()
```

```python
# Creating a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plotting in each subplot
axes[0, 0].hist(college['Outstate'], bins=20)
axes[0, 0].set_title('Outstate')

axes[0, 1].hist(college['Personal'], bins=20)
axes[0, 1].set_title('Personal')

axes[1, 0].hist(college['Apps'], bins=20)
axes[1, 0].set_title('Apps')

axes[1, 1].hist(college['Enroll'], bins=20)
axes[1, 1].set_title('Enroll')

# Adjusting layout to prevent overlap
plt.tight_layout()

# Showing the plot
plt.show()
```
