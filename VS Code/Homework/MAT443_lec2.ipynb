```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import statsmodels.api as sm
```

```python
# Load the Boston dataset from statsmodels
boston = sm.datasets.get_rdataset('Boston', 'MASS').data
###you can also load data from ISLP
from ISLP import load_data
# boston=load_data("Boston")
print(boston.columns)
# Attach Boston data
X = boston[['lstat']]
y = boston['medv']
```

```python
# Perform linear regression on lstat
lm = LinearRegression()
lm1=lm.fit(X, y)
lm1.coef_
lm1=sm.OLS(y,X)
lm1.fit().summary()
```

```python
# Perform linear regression on lstat and age
X_multi = boston[['lstat', 'age']]
lm_multi = LinearRegression()
lm_multi.fit(X_multi, y)
```

```python
# Display the summary using statsmodels
X_multi_with_const = sm.add_constant(X_multi)
lm_multi_sm = sm.OLS(y, X_multi_with_const).fit()
print(lm_multi_sm.summary())
```

```python
# Load the Caravan dataset from ISLR using statsmodels
caravan = sm.datasets.get_rdataset('Caravan', 'ISLR').data
#caravan=load_data("Caravan")
# Show first 30 rows
print(caravan.head(30))
```

```python
# Show dimensions
print(caravan.shape)
```

```python
# Summarize the 'Purchase' column
print(caravan['Purchase'].describe())
```

```python
# Scale the data excluding the 'Purchase' column
scaler = StandardScaler()
X_scaled = scaler.fit_transform(caravan.drop(columns=['Purchase']))
```

```python
# Split the data into train and test sets
test_indices = np.arange(0, 1000)
train_indices = np.arange(1000, caravan.shape[0])

X_train = X_scaled[train_indices]
X_test = X_scaled[test_indices]
y_train = caravan['Purchase'].iloc[train_indices]
y_test = caravan['Purchase'].iloc[test_indices]
```

```python
# Fit the K-NN model on the training data
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

# Predict on the training data
y_train_pred = knn.predict(X_train)

# Calculate the training accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy: {train_accuracy}")
```

```python
##confusion table
from ISLP import confusion_table
print(confusion_table(y_train_pred,y_train))
```

```python
knn_pred = knn.predict(X_test)
# Calculate the mean error rate
mean_error_rate = np.mean(y_test != knn_pred)
print(f"Mean error rate: {mean_error_rate}")
```

```python
# Show the confusion matrix
print(confusion_table(knn_pred,y_test))
```
